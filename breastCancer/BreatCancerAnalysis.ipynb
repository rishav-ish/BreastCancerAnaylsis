```python
from sklearn.linear_model import LogisticRegression
```

```python
sns.heatmap(confusion_matrix(ytest,prediction),annot=True,cmap='Blues')
```

# Plotting ROC Curve

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

```python
ypred = model.predict_proba(Xtest)[:,1]
```

```python
change = {
    'B':0,
    'M':1
}
```

```python
ytest = ytest.map(change)
```

```python
fpr, tpr, threshold = roc_curve(ytest,ypred)
```

```python
area = auc(fpr,tpr)
```

```python
plt.plot(fpr,tpr,c='r',label='area=%.2f'%area)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
```
